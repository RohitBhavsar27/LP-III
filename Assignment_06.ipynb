```python
#Converting the categorical columns.
productline = pd.get_dummies(df['PRODUCTLINE'])
dealsize = pd.get_dummies(df['DEALSIZE'])
```

```python
df = pd.concat([df,productline,dealsize], axis = 1)
```

```python
#Dropping the orderdate as month is already included
df.drop('ORDERDATE', axis = 1, inplace = True)
```

### Elbow plot to determine the number of clusters

```python
distortion = [] #Within Cluster sum of Squares from the centroid
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters = k)
    kmeanModel.fit(df)
    distortion.append(kmeanModel.inertia_) #Appending the inertia to the Distortion
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortion, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow method showing the optimal k')
plt.show()
```

#### As the number of k increases Inertia decreases.
#### Observation: A Elbow can be observed at 3 and after that the curve decreases gradually.

```python
# Returns a numpy array
X_train = df.values 
```

```python
X_train.shape
```

```python
model = KMeans(n_clusters = 3, random_state = 2) #Number of cluster = 3
model = model.fit(X_train) #Fitting the values to create a model
predictions = model.predict(X_train) #Predicting the cluster values (0,1 or 2)
```

```python
unique,counts = np.unique(predictions, return_counts = True)
```

```python
counts = counts.reshape(1,3)
```

```python
counts_df = pd.DataFrame(counts, columns = ['Cluster1','Cluster2','Cluster3'])
```

```python
counts_df.head()
```

### Visualization

```python
#Converting all the features into 2 columns to make it easy to visualize using PCA
pca = PCA(n_components = 2) 
```

```python
#Creating a DataFrame
reduced_X = pd.DataFrame(pca.fit_transform(X_train),columns = ['PCA1', 'PCA2'])
```

```python
reduced_X.head()
```

```python
#Plotting the normal Scatter Plot
plt.figure(figsize = (14,10))
plt.scatter(reduced_X['PCA1'],reduced_X['PCA2'])
```

```python
#Finding the centroids (3 centroid in total)
model.cluster_centers_
```

```python
#Transforming the centroid into 3 in x and y coordinate
reduced_centers = pca.transform(model.cluster_centers_)
```

```python
reduced_centers
```

```python
plt.figure(figsize = (14,10))
plt.scatter(reduced_X['PCA1'], reduced_X['PCA2'])
plt.scatter(reduced_centers[:,0],reduced_centers[:,1], color = 'black', marker = 'x', s = 300) #Plotting the centroid
```
